```python
import numpy as np
```

```python
import pandas as pd
```

```python
np.random.seed(0)

# Simulated data
temperature = np.linspace(300, 600, 30)        # K
yield_strength = 500 - 0.1*temperature + np.random.normal(0, 5, 30)

df = pd.DataFrame({
    "Temperature": temperature,
    "Yield_Strength": yield_strength
})
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x="Temperature",y="Yield_Strength",data=df)
```

```python
corr = df["Temperature"].corr(df["Yield_Strength"])
```

```python
print(f"correlation coeffient:{corr:.2f}")
```
